```python
import pandas as pd
import numpy as np
```

```python
#DATA LOADING
#this function should load the CSV file into pandas array
def readCSVTrain(url1):
    df = pd.read_csv(url1, header=None)
    return df

def readCSVTest(url2):
    df2 = pd.read_csv(url2, header=None)
    return df2

df = readCSVTrain("/Users/himadri/Desktop/CS Projects/train.csv")
df2 = readCSVTest("/Users/himadri/Desktop/HW3/test3.csv")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
#TRAINING PHASE
#spliting the training into features and targets
td=df.iloc[:,0:57]     #split into features
tr=df.iloc[:,-1]       #split into targets (0 or 1)

#just to check outputs
print (td)
print (tr)

#learned about this code using scikitlearn documentation 
#allows us to split our train set and test classifier on it
X_train, X_test, y_train, y_test = train_test_split(td, tr, test_size=0.333, random_state=0)
```

```python
#FIT CLASSIFIER
#implement the tree classifier
rfClassifier = RandomForestClassifier(n_estimators=200, max_depth=20, random_state=0)
rfClassifier.fit(X_train, y_train)

#got help from scikit learn tutorial for tree classifier score 
print ('training subset accuracy: {:.3f}'.format(rfClassifier.score(X_train, y_train)))
print ('testing subset accuracy: {:.3f}'.format(rfClassifier.score(X_test, y_test)))

importances = rfClassifier.feature_importances_
print (importances)

print (rfClassifier.n_features_)
```

```python
#PREDICT FOR THE TEST SET
dfFinal = readCSVTest("/Users/himadri/Desktop/HW3/test3.csv")

def predictFinal(classifier_name, test_dataframe):
    tr_predictFinal = classifier_name.predict(test_dataframe)
    return tr_predictFinal

#append the final row from predictor and write to csv
def appendCol(predictor, resultURL):
    dfFinal['57'] = predictor
    dfFinal.to_csv(resultURL, header=False)
    
col = predictFinal(rfClassifier, dfFinal)
appendCol(col, "/Users/himadri/Desktop/HW3/F002W5G_test3_result.csv")
```
